```python
import pandas as pd
import numpy as np
import utils as u
import my_ds_babel 
import csv
from io import StringIO
```

```python
df1=pd.read_csv("only_wood_customer_us_1.csv")
df2=pd.read_csv("only_wood_customer_us_2.csv",delimiter=';',names=['Age','City',"Gender","Name","Email"])
df3=pd.read_csv("only_wood_customer_us_3.csv",delimiter='\t|,')# the | sign is used to specify that we going to use two delimiter \t and ,
```

```python
df1
```

```python
#Checking columns Table 1
u.check_string_column(df1["Gender"]) #'This column contains 982 dirty values' 0 and 1 must be changed into Male and Female
u.check_string_column(df1["FirstName"]) # 'This column contains 642 dirty values') we must replace \\ "" characters by emplty and make names started by upcase letter
u.check_string_column(df1["LastName"]) #'This column contains 753 dirty values') we must replace \\ "" characters by emplty and make names started by upcase letter
u.check_numerical_column(df1["Age"]) # Age columns is clean we just need to convert each entry to str as the specified by the work
u.check_string_column(df1["City"]) # 'This column contains 3839 dirty values' we should replace - and _ characters by empty  and make Cities started by upcase letter
u.check_string_column(df1["Country"]) # 'This column contains 6647 dirty values' we will set the standar country name sa USA
```

```python
#Clean Table 1 
df1["Gender"]=df1["Gender"].apply(u.clean_gender)
df1['FirstName']=df1["FirstName"].apply(u.clean_name)
df1['LastName']=df1["LastName"].apply(u.clean_name)
df1["Email"]=df1["Email"].str.lower()
df1["Age"]=df1["Age"].apply(str)
df1['City']=df1["City"].apply(u.clean_city)
df1['Country']="USA"
df1.drop(['UserName'],inplace=True,axis=1)
```

```python
#Checking columns Table 2
u.check_numerical_column(df2['Age']) # all the value are str values but for some of them we have to retrive some additional str as 'year,yo,years' and replace them with empty
u.check_string_column(df2["City"]) # This column contains 3922 dirty values we should replace - and _ characters by empty  and make Cities started by upcase letter 
u.check_string_column(df2['Gender']) #This column contains 997 dirty values 0 and 1 must be changed into Male and Female: Same for M and F
u.check_string_column(df2["Name"]) #This column contains 10000 dirty values because of the escape, we will use it to separate the firstName and the last Name . Neithereless we must replace \\ "" characters by emplty and make names started by upcase letter.
```

```python
#Clean Table 2
df2["Age"]=df2['Age'].apply(u.clean_age)
df2["City"]=df2["City"].apply(u.clean_city)
df2["Gender"]=df2["Gender"].apply(u.clean_gender)
df2[["FirstName","LastName"]]=df2["Name"].str.split(expand=True)
df2['FirstName']=df2['FirstName'].apply(u.clean_name)
df2['LastName']=df2['LastName'].apply(u.clean_name)
df2["Email"]=df2["Email"].str.lower()
df2['Country']="USA"
df2.drop(["Name"],inplace=True,axis=1)
```

```python
#Checking columns Table 3 
df3['Gender'].unique() # the values found will be replaced by Femal or Male
u.check_string_column(df3["Name"])# the prefix string_ will cutted for all names same for Email 
u.check_string_column(df3["Age"]) 
```

```python
#Clean Table 3
df3["Gender"]=df3['Gender'].apply(u.clean_gender_tab3)
df3[["FirstName","LastName"]]=df3["Name"].apply(u.replace_string_).str.split(expand=True)
df3['FirstName']=df3['FirstName'].apply(u.clean_name)
df3['LastName']=df3['LastName'].apply(u.clean_name)
df3["Email"]=df3["Email"].apply(u.replace_string_).str.lower()
df3["Age"]=df3['Age'].apply(u.clean_age)
df3["City"]=df3["City"].apply(u.replace_string_).apply(u.clean_city)
df3['Country']="USA"
df3.drop(["Name"],inplace=True,axis=1)
df3
```

```python
df_merged=pd.concat([df1,df2,df3])
```
